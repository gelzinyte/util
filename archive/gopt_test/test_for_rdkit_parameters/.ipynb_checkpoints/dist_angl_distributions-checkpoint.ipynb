```python
import util
import matplotlib.pyplot as plt
from ase.io import read
import numpy as np
```

```python
ats_nice = read('rdkit_starts_CCCC_regular.xyz', ':')
ats_messy = read('rdkit_starts_CCCC.xyz', ':')
```

```python
nice_data = util.distances_dict(ats_nice)
messy_data = util.distances_dict(ats_nice)
```

```python
for xx in nice_data.keys():
    plt.figure()
    data = messy_data[xx]
    data = [d for d in data if d<2]
    plt.hist(data,  color='k')
    data = nice_data[xx]
    data = [d for d in data if d<2]
    plt.hist(data,  color='yellow', alpha=0.5)
    plt.title(xx)
    plt.xlabel('distance, AA')
    plt.show()
```
